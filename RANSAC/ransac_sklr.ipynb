# RANSAC с помощью sklern

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
def generate_random_x_y(k = 5, b = 20):
    x = np.arange(-100,100,1) 
    y = x*k+b + np.random.randint(-80,80, 200)   

    x = np.append(x,np.random.randint(-100,100,60)) 
    y = np.append(y,np.random.randint(-400,400,60))
    return x,y
```

```python
x,y = generate_random_x_y()
```

```python

plt.plot(x,y, 'r.')
```

```python
print(f"x size = {x.shape[0]} y size {y.shape[0]}")
```

```python
ransac_model = linear_model.RANSACRegressor()
ransac_model.fit(x.reshape(x.shape[0],1),y)
```

```python
x_pr = np.arange(-50,50,1) + np.random.randint(-100,100,100)
y_pr = ransac_model.predict(x_pr.reshape(x_pr.shape[0],1))
```

```python
plt.plot(x_pr, y_pr, 'b')
plt.plot(x, y, '.r')
```
